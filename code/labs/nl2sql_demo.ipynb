```python
import duckdb
from langchain_community.utilities import SQLDatabase
```

```python
from dotenv import load_dotenv
import os
# Load environment variables from .env file
load_dotenv()
```

## Connect database

```python
#db = duckdb.connect("cfr2sbvr_db/database_v3.db")

db = SQLDatabase.from_uri("duckdb:///cfr2sbvr_db/database_v3.db")
```

Test connectivity

```python
query = """
show tables;
"""

tables = db.run(query)

tables
```

```python
from langchain_openai import OpenAI
from langchain_community.agent_toolkits import create_sql_agent

llm = OpenAI(temperature=0)  # Make sure your OpenAI API key is configured

agent_executor = create_sql_agent(llm, db=db, verbose=True)
```

```python
agent_executor.invoke({"input": "Checkpoints are file_source in many tables. For tranformation, list all checkpoints?"})
```

```python
agent_executor.invoke({"input": "How many terms are per checkpoint?"})
```

```python
agent_executor.invoke({"input": 
"""
Transform tables has all elements transformed (terms, names, operative rules, and fact types)
The content.statement_name has the term name, and file_source has the checkpoint name.
How many terms are per checkpoint?
"""
})
```
